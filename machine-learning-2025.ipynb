### Exercise 1: Splitting and Preparing the Data

* Convert it into a pandas DataFrame and check for missing values.

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()

# convert data into a pandas Dataframe
diabetes_df = pd.DataFrame(diabetes.data,
                          columns=diabetes.feature_names)

diabetes_df["target"] = diabetes.target

# check for missing values
diabetes_df.isnull().sum()
```

```python
# preparing dataset / split data

from sklearn.model_selection import train_test_split
X = diabetes_df.dropna()
X = diabetes_df.drop("target", axis=1)
y = diabetes_df["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

### Exercise 3: Choosing and Training a Model
* Based on whether the problem is classification or regression, choose an appropriate model (e.g., LogisticRegression, RandomForestClassifier, LinearRegression, etc.).
* Train the model using the training data.
* Print the model parameters after training.

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()

# fit the data
model.fit(X_train, y_train)
```

```python
# make predictions
y_preds = model.predict(X_test)

# check score
model.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae = mean_absolute_error(y_test, y_preds)
mse = mean_squared_error(y_test, y_preds)
rmse = np.sqrt(mse)

print(f"mae: {mae:.2f}, mse: {mse:.2f}, rmse: {rmse:.2f}")
```
